# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[:10]
```

```python
income_data[:10]
```

```python
#N(m,s) -> m + s * Z
height = 160 + 30 * np.around(np.random.randn(n),1)
weight = 65 + 25 * np.around(np.random.randn(n),1)
age = 30 + 12 * np.around(np.random.randn(n),0)
income = 18000 + 3500 * np.around(np.random.randn(n),0)
```

```python
data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income
    }
)
```

```python
data.head()
```

```python
%matplotlib inline
plt.hist(data["Height"])
```

## Agrupación de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

## Doble agrupación

```python
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

# Operaciones sobre datos agrupados

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
grouped_income = double_group["Income"]
```

```python
grouped_income.describe()
```

```python
double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": np.std
    }
)
```

## Utilizar una función (lambda) personalizada para agregar datos

```python
double_group.aggregate(
    {
        "Age": np.mean,
        "Height": lambda h:np.mean(h)/np.std(h)
    }
)
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x)/ np.std(x)])
```

# Filtrado de datos

```python
double_group["Age"].filter(lambda x: x.sum()>2400)
```

## Transformación de variables

```python
zscore = lambda x: (x - np.mean(x))/np.std(x)
```

```python
double_group.transform(zscore)
```

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
