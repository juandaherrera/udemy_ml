# Clustering jerárquico y dendrogramas

Antes de empezar, pongamos un poco de notación para hablar todos el mismo idioma

* X dataset (array de n x m) de puntos a clusterizar
* n número de datos
* m número de rasgos
* Z array de enlace del cluster con la información de las uniones
* k número de clusters

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0], [[3,1], [1,4]], size = [100,])
b = np.random.multivariate_normal([0,20], [[3,1], [1,4]], size = [50,])

X = np.concatenate((a,b))
print(X.shape)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
Z = linkage(X, 'ward')
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
```

El coeficiente de cofenética es un número entre -1 y 1, donde un valor más cercano a 1 indica una mayor correlación entre las distancias originales y las distancias en el dendrograma, lo que sugiere que el agrupamiento jerárquico preserva de manera efectiva las distancias originales entre los puntos de datos. Por otro lado, un coeficiente de cofenética cercano a 0 o negativo indica que el dendrograma no preserva bien las distancias originales entre los puntos de datos.

```python
c, coph_dist = cophenet(Z, pdist(X))
c
```

### Truncar el dendograma|

```python
plt.figure(figsize=(25,10))
plt.title('Dendrograma del clusterin jerárquico TRUNCADO')
plt.xlabel('Índices de la Muestra')
plt.ylabel('Distancias')
dendrogram(Z, leaf_rotation=90., leaf_font_size=12.0, 
           truncate_mode='lastp', p=10, show_leaf_counts=True, #los ultimos 10 clusters unidos (p=10)
          show_contracted=True) 
plt.show()
```

### Dendograma tuneado

```python
def dendrogram_tune(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
        
    annotate_above = kwargs.pop('annotate_above', 0)
    
    ddata = dendrogram(*args, **kwargs)
    
    if not kwargs.get('no_plot', False):
        plt.title('Clustering jerárquico con Dendograma truncado')
        plt.xlabel('Índice del Dataset (o tamaño del cluster)')
        plt.ylabel('Distancia')
        
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x,y,'o', c=c)
                plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),
                            textcoords='offset points', va='top', ha='center')
    if max_d:
        plt.axhline(y=max_d, c='k')
        
    return ddata
```

```python
dendrogram_tune(Z, truncate_mode='lastp', p=12, leaf_rotation=90.,
               leaf_font_size=12,show_contracted=True, annotate_above = 10)
```

## Corte automático del dendrograma

inconsistency_i = (h_i-avg(h_j))/std(h_j)

```python
from scipy.cluster.hierarchy import inconsistent
```

```python
depth = 5
incons = inconsistent(Z, depth)
incons[-10:]
```

## Método del codo

```python
last = Z[-10:, 2] #los últimos 10y solo la columna 2
last_rev = last[::-1] # Cambiar el órden (los pone al contrario de como estaban)
print(last_rev)
idx = np.arange(1, len(last)+1)
plt.plot(idx, last_rev)

acc = np.diff(last, 2)
acc_rev = acc[::-1]
plt.plot(idx[:-2] + 1, acc_rev)
plt.show()
k = acc_rev.argmax() + 2
print("El numero de optimo de cluster es %s"%str(k))
```
